```python
# Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#Import the dataset
data = pd.read_csv('stock1.csv')
```

```python
data.head()
```

```python
new_data = data.drop(0)
```

```python
new_data.head()
```

```python
# Convert 'Date' to datetime and set as index
new_data['Date'] = pd.to_datetime(new_data['Date'])
new_data['Date'] = new_data['Date'].dt.date
new_data.set_index('Date', inplace=True)
```

```python
new_data.head()
```

```python
# Convert 'Close' column to numeric, coercing errors to NaN
new_data['Close'] = pd.to_numeric(new_data['Close'], errors='coerce')
new_data['Close']
```

```python
# To make a plot on plotly
fig = px.line(new_data, x=new_data.index, y='Close', title='Close Price History')
fig.show()
```

```python
new_data['Close_lastMonth'] = new_data['Close'].shift(+1)
new_data['Close_last2Month'] = new_data['Close'].shift(+2)
new_data['Close_last3Month'] = new_data['Close'].shift(+3)
new_data['Close_ last4Month'] = new_data['Close'].shift(+4)
new_data['Close_last5Month'] = new_data['Close'].shift(+5)
new_data['Close_last6Month'] = new_data['Close'].shift(+6)
new_data
```

```python
new_data1 = new_data[['Close', 'Close_lastMonth', 'Close_last2Month', 'Close_last3Month', 'Close_ last4Month', 'Close_last5Month', 'Close_last6Month']]
```

```python
new_data1.tail(365)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lin_model = LinearRegression()
rf_model = RandomForestRegressor(n_estimators=100,max_features=7,random_state=1)
svr_model = SVR(kernel='rbf')
dt_model = DecisionTreeRegressor(random_state=1)
```

```python
x1,x2,x3,x4,x5,x6,y = new_data1['Close_lastMonth'],new_data1['Close_last2Month'],new_data1['Close_last3Month'],new_data1['Close_ last4Month'],new_data1['Close_last5Month'],new_data1['Close_last6Month'],new_data1['Close']
x1,x2,x3,x4,x5,x6,y = np.array(x1),np.array(x2),np.array(x3),np.array(x4),np.array(x5),np.array(x6),np.array(y)
x1,x2,x3,x4,x5,x6,y = x1.reshape(-1,1),x2.reshape(-1,1),x3.reshape(-1,1),x4.reshape(-1,1),x5.reshape(-1,1),x6.reshape(-1,1),y.reshape(-1,1)
final_x = np.concatenate((x1,x2,x3,x4,x5,x6),axis=1)
final_x
```

```python
# Drop rows with NaN values
cleaned_x = final_x[~np.isnan(final_x).any(axis=1)]
cleaned_y = y[~np.isnan(final_x).any(axis=1)]

#TO SPLIT THE DATASET
x_train,x_test,y_train,y_test = train_test_split(cleaned_x,cleaned_y,test_size=0.2,random_state=1)
```

```python
model1 = lin_model.fit(x_train,y_train)
model2 = rf_model.fit(x_train,y_train)
model3 = svr_model.fit(x_train,y_train)
model4 = dt_model.fit(x_train,y_train)
```

```python
pred1 = model1.predict(x_test)
pred2 = model2.predict(x_test)
pred3 = model3.predict(x_test)
pred4 = model4.predict(x_test)

```

```python
#To plot actual vs predicted
import pandas as pd
temp_df = pd.DataFrame({'Actual': y_test.flatten(), 'Linear Regression': pred1.flatten(), 'Random Forest Regression': pred2.flatten(),'Support Vector Regression': pred3.flatten(),'Decision Tree Regression': pred4.flatten()})
linear = px.line(temp_df, y=['Actual', 'Linear Regression'], title='Actual vs Predicted')
random_forest = px.line(temp_df, y=['Actual', 'Random Forest Regression'], title='Actual vs Predicted')
support_vector = px.line(temp_df, y=['Actual', 'Support Vector Regression'], title='Actual vs Predicted')
decision_tree = px.line(temp_df, y=['Actual', 'Decision Tree Regression'], title='Actual vs Predicted')
linear.show()
random_forest.show()
support_vector.show()
decision_tree.show()
```

```python
mse = mean_squared_error(y_test, pred1)
r2 = r2_score(y_test, pred1)
print("MSE:", mse)
rmse = np.sqrt(mean_squared_error(y_test, pred1))
r2 = r2_score(y_test, pred1)
print("RMSE:", rmse)
print("R2 Score:", r2)
```
